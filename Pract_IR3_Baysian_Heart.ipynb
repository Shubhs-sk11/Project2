```python
import numpy as np
import pandas as pd
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
```

```python
# Read Cleveland Heart Disease data 
heartDisease = pd.read_csv("heart.csv")
```

```python
heartDisease = heartDisease.replace('?', np.nan)
```

```python
# Model Bayesian Network
model = DiscreteBayesianNetwork([('age','trestbps'),('age','fbs'),('sex','trestbps'),('exang','trestbps'),('trestbps','heartdisease'),
                       ('fbs','heartdisease'),('heartdisease','restecg'), ('heartdisease','thalach'),('heartdisease','chol')])
```

```python
# Learning CPDs using Maximum Likelihood Estimators
print('\nLearning CPD using Maximum likelihood estimators')
model.fit(heartDisease, estimator=MaximumLikelihoodEstimator)
```

```python
# Inferencing with Bayesian Network 
print('Inferencing with Bayesian Network:') 
HeartDisease_infer = VariableElimination(model)

# Probability of HeartDisease given Age = 38
print('1. Probability of HeartDisease given Age=29') 
q1 = HeartDisease_infer.query(variables=['heartdisease'], evidence={'age':29})
print(q1)

# Probability of HeartDisease given cholesterol 
print('\n 2. Probability of HeartDisease given cholesterol=126') 
q=HeartDisease_infer.query(variables=['heartdisease'], evidence ={'chol':126})
print(q)
```
