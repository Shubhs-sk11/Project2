```python
import io
import nltk
nltk.download("punkt")
nltk.download("stopwords")

```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
stop_words = set(stopwords.words('english'))
ps = PorterStemmer()
```

```python
with open("ssk_text.txt", "w") as f:
    f.write("This is an example sentence to test the stop words filtration.")
# Read original file
file1 = open("ssk_text.txt")
line = file1.read()
```

```python
print(line)
```

```python
# Split into words (YOUR ORIGINAL LOGIC)
words = line.split()
```

```python
print(words)
```

```python
print(f"Before Removing Stop Words : \n{words}\n Length : {len(words)}")
```

```python
with open("ssk_filtered_text.txt", "w") as appendFile:
    for r in words:
        if not r.lower() in stop_words:
            appendFile.write(" " + r)
```

```python
# Read filtered result from file (YOUR FLOW)
file2 = open("ssk_filtered_text.txt")
line2 = file2.read()
filtered_words = line2.split()
```

```python
print(f"\nAfter Removing Stop Words : \n{filtered_words}\nLength : {len(filtered_words)}")
```

```python
import string
stemmed_list = []

for w in filtered_words:
    # remove punctuation like '.', ',', etc.
    cleaned_word = w.strip(string.punctuation)

    # stemming
    stemmed_word = ps.stem(cleaned_word)

    stemmed_list.append(stemmed_word)
```

```python
print(f"\nAfter Stemming : \n{stemmed_list}\nLength : {len(stemmed_list)}")
```
